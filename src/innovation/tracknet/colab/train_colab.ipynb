```python
# Step 1: Verify GPU
import torch
print('CUDA available:', torch.cuda.is_available())
if torch.cuda.is_available():
    print('GPU:', torch.cuda.get_device_name(0))
else:
    print('WARNING: No GPU detected. Training will be very slow on CPU.')
```

```python
# Step 2: Verify all required files are present
import os
required = [
    'tracknet_model.py',
    'tracknet_dataset.py',
    'tracknet_train.py',
    'ball_labels.csv',
    'men360.mp4',
]
all_ok = True
for f in required:
    path = f'/content/{f}'
    exists = os.path.exists(path)
    size   = f'  ({os.path.getsize(path):,} bytes)' if exists else ''
    status = 'OK     ' if exists else 'MISSING'
    print(f'{status}  {f}{size}')
    if not exists:
        all_ok = False
if all_ok:
    print('\nAll files present. Ready to train.')
else:
    print('\nUpload the missing files before continuing.')
```

```python
# Step 4: Train
# All paths are already set to /content/ inside tracknet_train.py
import sys
sys.path.insert(0, '/content')

from tracknet_train import train

train(
    label_path = '/content/ball_labels.csv',
    video_dir  = '/content',
    out_dir    = '/content/weights',
    epochs     = 30,
    batch_size = 4,
    lr         = 5e-4,
)
```

```python
# Step 5: Confirm saved weights
import os
weights_dir = '/content/weights'
if os.path.exists(weights_dir):
    for f in sorted(os.listdir(weights_dir)):
        size = os.path.getsize(os.path.join(weights_dir, f))
        print(f'{f}  ({size:,} bytes)')
else:
    print('No weights folder found — did training complete?')
```
